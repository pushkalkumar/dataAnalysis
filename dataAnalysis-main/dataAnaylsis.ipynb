```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'Air_Quality.csv' 
df = pd.read_csv(file_path)
```

```python
# Pollutants categories based on the 'Name' column 
pollutants = {
    "SO2": "Boiler Emissions- Total SO2 Emissions",
    "PM2.5": "Fine particles (PM 2.5)",
    "NO2": "Nitrogen dioxide (NO2)",
    "Asthma_PM2.5": "Asthma emergency department visits due to PM2.5",
    "VMT": "Annual vehicle miles traveled",
    "Resp_PM2.5": "Respiratory hospitalizations due to PM2.5 (age 20+)",
    "O3": "Ozone (O3)",
    "Asthma_O3": "Asthma hospitalizations due to Ozone",
    "Deaths_O3": "Cardiac and respiratory deaths due to Ozone",
    "NOx": "Boiler Emissions- Total NOx Emissions",
    "Cardio_PM2.5": "Cardiovascular hospitalizations due to PM2.5 (age 40+)",
    "Formaldehyde": "Outdoor Air Toxics - Formaldehyde",
    "VMT_Cars": "Annual vehicle miles traveled (cars)",
    "VMT_Trucks": "Annual vehicle miles traveled (trucks)"
}

# Dictionary to store the filtered data for each pollutant
filtered_data = {key: [] for key in pollutants.keys()}

# Filter dataset by pollutant 
for key, pollutant in pollutants.items():
    filtered_data[key] = df[df['Name'] == pollutant]

# Plot pollutant
def plot_pollutant(pollutant_key):
    data = filtered_data[pollutant_key]
    if data.empty:
        print(f"No data for {pollutant_key}")
        return
    
    # Sort the data by the time period
    data = data.sort_values(by='Time Period')
    data = data[data['Time Period'].str.contains("Annual Average")]
    # Plotting the Data Value against Time Period
    plt.figure(figsize=(15, 6))
    plt.bar(data['Time Period'], data['Data Value'], color='skyblue')
    plt.title(f"{pollutants[pollutant_key]} Over Time")
    plt.xlabel("Time Period")
    plt.ylabel("Data Value")
    plt.xticks(rotation=90)
    plt.xticks(data['Time Period'][::2], rotation =90)
    plt.tight_layout()
    plt.show()
```

```python
#Plotting PM 2.5
plot_pollutant("PM2.5")
```

```python
#Plotting Nitrogen dioxide (NO2)
plot_pollutant("NO2")
```
